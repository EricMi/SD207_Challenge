```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import librosa
from presets import Preset
from time import time
```

```python
# Define some usefull functions
def load_sound_files(file_paths):
    raw_sounds = []
    for fp in file_paths:
        X,sr = librosa.load(fp)
        raw_sounds.append(X)
    return raw_sounds
```

```python
# Read data and preprocessing
X_test = pd.read_csv('test_files.txt', header=None)[0].values
data = pd.read_csv('audio/train.txt', header=None, sep='\s+')
X_train =  data[0].values
y_train = pd.factorize(data[1])[0]
#data_val = pd.read_csv('audio/dev.txt', header=None, sep='\s+')
#X_train = data[0].values
print X_train
#print y_train
```

```python
n_mfcc = 20
def extract_feature(file_name):
    X, sample_rate = librosa.load(file_name, sr = None)
    stft = np.abs(librosa.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=20).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X),
    sr=sample_rate).T,axis=0)
    return mfccs,chroma,mel,contrast,tonnetz

def parse_audio_files(files, file_labels):
    features, labels = np.empty((0,173)), np.empty(0)
    for i in range(files.shape[0]):
        try:
            mfccs, chroma, mel, contrast,tonnetz = extract_feature(files[i])
            #print mfccs.shape,chroma.shape,mel.shape,contrast.shape,tonnetz.shape
        except Exception as e:
            print "Error encountered while parsing file: ", files[i]
            continue
        ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
        #print ext_features.shape
        features = np.vstack([features,ext_features])
        labels = np.append(labels, file_labels[i])
    return np.array(features), np.array(labels, dtype = np.int)
```

```python
ft, lb = parse_audio_files(X_train, y_train)
print ft.shape
print lb
```

```python
def parse_audio_test_files(files):
    features = np.empty((0,173))
    for i in range(files.shape[0]):
        print files[i]
        try:
            mfccs, chroma, mel, contrast,tonnetz = extract_feature(files[i])
            #print mfccs.shape,chroma.shape,mel.shape,contrast.shape,tonnetz.shape
        except Exception as e:
            print "Error encountered while parsing file: ", files[i]
            continue
        ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
        #print ext_features.shape
        features = np.vstack([features,ext_features])
    return np.array(features)
```

```python
X = ft
y = lb

Xt = parse_audio_test_files(X_test)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(173,))
clf.fit(X, y)

y_pred = clf.predict(Xt)
```
